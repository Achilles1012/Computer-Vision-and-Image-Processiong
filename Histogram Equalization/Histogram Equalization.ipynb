```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
img = cv2.imread("C:/Users/Achilles2000/Desktop/Semester 2/CV/Assignments/Histogram Equalization/Elk.jpg")
# Display Input image
Image(filename='C:/Users/Achilles2000/Desktop/Semester 2/CV/Assignments/Histogram Equalization/Elk.jpg')
```

```python

# Convert BGR to YCbCr
yCbCr_img = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

# Extract Y channel
Y_channel = yCbCr_img[:,:,0]

# Display histogram of Y channel
plt.hist(Y_channel.ravel(),256,[0,256])
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('Histogram of Y channel (Before Histogram Equalization)')
plt.show()

# Apply histogram equalization on Y channel
equ_Y_channel = cv2.equalizeHist(Y_channel)

# Update Y channel in YCbCr image
yCbCr_img[:,:,0] = equ_Y_channel

# Display histogram of equalized Y channel
plt.hist(equ_Y_channel.ravel(),256,[0,256])
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('Histogram of Y channel (After Histogram Equalization)')
plt.show()

# Convert YCbCr back to BGR
output_img = cv2.cvtColor(yCbCr_img, cv2.COLOR_YCrCb2BGR)
cv2.imwrite("ElkEnhanced.jpg", output_img)
# Display output image
#cv2.imshow('Output Image', output_img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

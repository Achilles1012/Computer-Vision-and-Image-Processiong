```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image = cv2.imread('cr7.jpg', 0)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()

# Add Gaussian noise
mean = 50
var = 100
sigma = var**0.5
gaussian = np.random.normal(mean, sigma, image.shape)
noisy_image = image + gaussian
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)  # Clip values to [0, 255]
plt.imshow(noisy_image, cmap='gray')
plt.title('Noisy Image')
plt.axis('off')
plt.show()

# Define a mean kernel
mean_filter = np.ones((3, 3)) / 9

# Apply mean filter using filter2D
mean_filtered_image = cv2.filter2D(noisy_image, -1, mean_filter)
plt.imshow(mean_filtered_image, cmap='gray')
plt.title('Mean Filtered Image')
plt.axis('off')
plt.show()

# Apply median filter to remove noise
median_filtered_image = cv2.medianBlur(noisy_image, 5)  # Adjust kernel size as needed
plt.imshow(median_filtered_image, cmap='gray')
plt.title('Median Filtered Image')
plt.axis('off')
plt.show()
```
